```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Position_Salaries.csv')
df.info()
```

```python
# as we dont need other pre processing steps 
# splitting the data in to train and test data set

X=df.iloc[:,1:2].values
y=df.iloc[:,2].values
y
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(random_state=0)
```

```python
regressor.fit(X,y)
```

```python
y_pred=regressor.predict([[6.5]])
y_pred
```

```python
# vislualising the data 
plt.scatter(X,y)
plt.plot(X,regressor.predict(X),c='r')
plt.xlabel("level")
plt.ylabel("salary")
plt.title("Decision tree regression ")
plt.show()
```

**The above plot is some kind of bluff to see the result of Decison tree regression model we have to look for high resolution model this is not a linear  model as well non linear model . it is continuous non linear model**

```python
X_grid=np.arange(min(X),max(X),0.01)
X_grid=X_grid.reshape((len(X_grid),1))
plt.scatter(X,y)
plt.plot(X_grid,regressor.predict(X_grid),c='r')
plt.xlabel('level')
plt.ylabel("salary")
plt.title("Decision tree regression high resolution")
plt.show()
```
